## Deteksi OUTLIER() 

- Pengertian Deteksi OUTLIER()

Suatu metode yang digunakan dalam analisis data untuk mengidentifikasi titik-titik data yang dianggap sebagai outlier atau anomali. Outlier adalah nilai yang secara signifikan berbeda dari sebagian besar data lainnya dalam sebuah dataset. Outlier dapat mengindikasikan kesalahan dalam pengumpulan data, variasi alami yang ekstrim, atau kejadian yang tidak biasa.

K-distance (dist_K) adalah jarak ke tetangga terdekat ke K. K-neighbors dilambahkan dengan Nk(A) mencakup himpunan titik yang terletak di dalam atau pada lingkaran berjari-jari jarak K. K-tetangga bisa lebih dari atau sama dengan nilai K.

Rumus hitung jarak Manhattan

$$d(x,y) = \sqrt{\sum_{i=1}^{n}(x_i - y_i)}$$

- contoh proses manual (Data berbeda dengan 3 fitur (3 dimensi) dengan jumlah data terserah)

- implementasikan ke Python

### Perhitungan LOF menggunakan library Scikit 

```
from sklearn.neighbors import LocalOutlierFactor

X = [
    (40,	75,	87),
    (60,	95,	56),
    (80,	67,	50),
    (50,	56,	86),
    (40,	87,	56),
    (60,	94,	78),
    (80,	67,	98),
    (70,	45,	56),
    (45,	78,	45),
    (85,	57,	80)

]

lof = LocalOutlierFactor(n_neighbors=2, contamination=0.1)

y_pred = lof.fit_predict(X)

print("Predicted table:", y_pred)
print("Negative LOF scores:", -lof.negative_outlier_factor_)  # Ubah tanda negatif untuk mendapatkan nilai LOF positif
```

## Naive Bayes Classiefier

- Naive Bayes Classifier
1. Konsep utama naive bayes classifier adalah probabilitas yang paling besar
 > data yang dipakai harus klasifikasi
2. Klasifikasi
 > memprediksi label class (diskrit atau continue)
	- klasifikasi yaitu daata yang berlabel bisa kategori atau numerik
	- cluster (prediksi) yaitu data yang tidak berlabel 
 > mengklasifikasi data (membangun model)
	- pada data training dan nilai label class dalam mengklasifikasikan atribut dan digunakannya pada saat mengklasifikasikan data baru 

      3. prediksi
       > memodelkan fungsi bernilai kontinue (nilai yang tidak diketahui)

      4. Data Preparation
      > Data cleaning
         - Memproses awal  data untuk mengurangi noise dan mengatasi nilai-nilai yang hilang
         - Analisa relevansi (seleksi fitur)
         - Menghilangkan atribut-atribut yang tidak relevan atau atribut yang redundan
      > Transformasi data
         - Membangun normalisasi data

-  Evaluting Classification Methods
> akurasi
  - keakuratan klasifikasi (memperkirakan label class)
  - keakurasian prediksi (nilai yang ditebak dari atribut yang diprediksi)
> kecepatan
  - waktu untuk membangun model (training time)
  - waktu dalam menggunakan model

- Rumus Naive Bayes Gaussian

$$P(H|A) = \frac{P(A|H) \cdot P(H)}{P(A)}$$

- Hitunglah manual sklearn dan implementasi menggunakan naive bayes GaussianNB as np

pada program dipecah data training (yang dihitung manual)

- setiap anak harus beda angka prediksi yang dihitung

### Program Perhitungan Gaussian Naive bayes 

```
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import csv

# Load the iris dataset
iris = load_iris()
X = iris.data
y = iris.target

# Initialize the Naive Bayes classifier
nb_classifier = GaussianNB()

# Split the data into training and test sets (80% training, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Train the classifier
nb_classifier.fit(X_train, y_train)

# Predict the classes for the training set
y_test_pred = nb_classifier.predict(X_test)

# Combine features and target into one list
train_data = np.column_stack((X_train, y_train))

# Write the training data to a CSV file
with open('training_data.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    # Write header
    writer.writerow(iris.feature_names + ['Target'])
    # Write data
    writer.writerows(train_data)

# Calculate the accuracy of the classifier on the training set
training_accuracy = accuracy_score(y_test, y_test_pred)
print("Training Accuracy:", training_accuracy)

# Sample data for prediction
sample = np.array([[5.3, 3.4, 4.2, 1.1]])

# Predict the class for the sample
prediction = nb_classifier.predict(sample)
print('Prediction:', prediction)
```
